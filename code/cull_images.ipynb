```python
import os
from matplotlib import pyplot as plt
import re
```

```python
def cull_images(items, origin_dir, discard_dir):
# items is the list of directories storing images of each item
    img = None
#     curr_path = os.path.abspath(__file__)
    curr_path = os.getcwd() + '/'
    fig = plt.figure(1)
    plt.ion()
    from IPython.core.debugger import Tracer
#     Tracer()()
    for item in items:
        path = curr_path + origin_dir + '/' + item + '/'
        discard_path = curr_path + discard_dir + '/' + item + '/'
        if not os.path.exists(discard_path):
            os.makedirs(discard_path)
        for d in os.listdir(path):
#             Tracer()()
            print d
            if d[0] == '.': # handle .DS_store
                continue
            this_d = path + d + '/'
            for img_f in os.listdir(this_d):
#                 Tracer()()
                print img_f
                this_f = this_d + img_f
                im=plt.imread(this_f)
#                 Tracer()()
                if img is None:
                    img = plt.imshow(im)
                else:
                    img.set_data(im)
                plt.pause(.1)
            #     pl.draw()
#                 Tracer()()
                instruction = raw_input("What do you want? ")
                if instruction == 'n':
                    os.rename(this_f, discard_path + img_f)
                elif instruction == 'q':
                    break
    plt.close(fig)
```
